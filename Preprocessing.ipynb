# **PRE-PROCESSING**

We use **image_dataset_from_directory** method for resizing and grayscaling

***We are resizing images as 256x256***

---



```python
data = tf.keras.utils.image_dataset_from_directory("/content/drive/MyDrive/data/data", color_mode='grayscale', image_size=(256, 256))
```

### **Normalization and deciding the train, val and test sizes**

```python
data =  data.map(lambda x,y: (x/255, y))
```

```python
#check the data length
len(data) 
```

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1) + 1
```

```python
#check if data length is provided
train_size + val_size + test_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```
